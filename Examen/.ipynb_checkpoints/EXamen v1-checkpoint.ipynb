```python
from neo4j import GraphDatabase

class Neo4jService(object):

    def __init__(self, uri, user, password):
        self._driver = GraphDatabase.driver(uri, auth=(user, password))

    def close(self):
        self._driver.close()      
    
    def crear_Alcaldes(self, tx, nombre):
        tx.run("CREATE (:Alcaldes {nombre: $nombre})", nombre=nombre)
    
    def crear_Alcalde(self, tx, nombre):
        tx.run("CREATE (:Alcalde {nombre: $nombre})", nombre=nombre)
        
    def crear_Titulo(self, tx, nombre):
        tx.run("CREATE (:Titulos {nombre: $nombre})", nombre=nombre)     
        
    def crear_Noticia(self, tx, nombre):
        tx.run("CREATE (:Noticias {nombre: $nombre})", nombre=nombre)        
        
    def crear_Fecha(self, tx, nombre):
        tx.run("CREATE (:Fechas {nombre: $nombre})", nombre=nombre)
            
    def crear_Link(self, tx, nombre):
        tx.run("CREATE (:Links {nombre: $nombre})", nombre=nombre)
    
    #realciones central - alcandes
    def crear_relacion_Alcaldes(self, tx, nombre_alcalde, nombre_alcaldes):
        tx.run("MATCH (a:Alcaldes {nombre: $nombre_alcalde}) "
               "MATCH (b:Alcalde {nombre: $nombre_alcaldes}) "
               "MERGE (a)-[:Alcaldes]->(b)",
               nombre_alcalde=nombre_alcalde, nombre_alcaldes=nombre_alcaldes)
           
#realciones alcalde - titulo
    def crear_relacion_AlTit(self, tx, nombre_alcalde, nombre_titulo):
        tx.run("MATCH (a:Alcalde {nombre: $nombre_alcalde}) "
               "MATCH (b:Titulos {nombre: $nombre_titulo}) "
               "MERGE (a)-[:Alcalde_Titulo]->(b)",
               nombre_alcalde=nombre_alcalde, nombre_titulo=nombre_titulo)
        
#realciones titulo - noticia 
    def crear_relacion_TitNot(self, tx, nombre_titulo,nombre_noticia):
        tx.run("MATCH (a:Titulos {nombre: $nombre_titulo}) "
               "MATCH (b:Noticias {nombre: $nombre_noticia}) "
               "MERGE (a)-[:Titulo_Noticia]->(b)",
               nombre_titulo=nombre_titulo, nombre_noticia=nombre_noticia)
#realciones titulo - fecha 
    def crear_relacion_TitFech(self, tx, nombre_titulo,nombre_fecha):
        tx.run("MATCH (a:Titulos {nombre: $nombre_titulo}) "
               "MATCH (b:Noticias {nombre: $nombre_fecha}) "
               "MERGE (a)-[:Titulo_Fecha]->(b)",
               nombre_titulo=nombre_titulo, nombre_fecha=nombre_fecha)
#realciones titulo - link 
    def crear_relacion_TitLink(self, tx, nombre_titulo,nombre_link):
        tx.run("MATCH (a:Titulos {nombre: $nombre_titulo}) "
               "MATCH (b:Noticias {nombre: $nombre_link}) "
               "MERGE (a)-[:Titulo_Link]->(b)",
               nombre_titulo=nombre_titulo, nombre_link=nombre_link)
        

    
print("Ejecucion correcta")
```
