```python
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox as MessageBox
import sqlite3

con = sqlite3.connect('mibase.db')
cursorObj = con.cursor()

def tablas(con):

    con.execute('''CREATE TABLE cliente
             (nombre     TEXT    NOT NULL,
                 direccion     TEXT     NOT NULL,
                 telefono      REAL,
                 email         TEXT     NOT NULL;''')
    con.execute('''CREATE TABLE cuenta
             (saldo     REAL,
                 tarjeta     REAL,
                 nombre      TEXT NOT NULL;''')
    
    con.execute('''CREATE TABLE producto
             (nombre     TEXT    NOT NULL,
                 precio     REAL,
                 stock      REAL
                 ;''')
    
class Aplicacion:
    def cliente(self):
        nombre=self.entry.get()
        direccion=self.entry1.get()
        telefono=self.entry2.get()
        email=self.entry3.get()
        cursorObj.execute('INSERT INTO cliente  VALUES (?, ?, ?, ?)', (nombre,direccion,telefono,email))
        con.commit()

    
    def cuenta(self):
        saldo=self.entry4.get()
        tarjeta=self.entry5.get()
        nombreC=self.entry6.get()
        cursorObj.execute('INSERT INTO cuenta  VALUES (?, ?, ?)', (saldo,tarjeta,nombreC))
        con.commit()
    
    def producto(self):
        nombre=self.entry14.get()
        precio=self.entry15.get()
        stock=self.entry16.get()
        cursorObj.execute('INSERT INTO producto  VALUES (?, ?, ?)', (nombre,precio,stock))
        con.commit()
    
    def Bcliente(self):
        nombre=self.entry17.get()
        direccion=self.entry18.get()
        telefono=self.entry19.get()
        email=self.entry20.get()
        cursor=cursorObj.execute("select nombre,direccion,telefono,email from cliente where nombre=?", (nombre, ))
        print(nombre+str(cursor))
        fila=cursor.fetchone()
        self.entry18.insert(0, fila[1])
        self.entry19.insert(0, fila[2])
        self.entry20.insert(0, fila[3])
        if fila!=None:
            print(fila)
        else:
            print("No existe.")

    def Bproducto(self):
        nombre=self.entry14.get()
        precio=self.entry15.get()
        stock=self.entry16.get()
        cursor=cursorObj.execute("select nombre,precio,stock from producto where nombre=?", (nombre, ))
        fila1=cursor.fetchone()
        self.entry22.insert(0, fila1[1])
        self.entry23.insert(0, fila1[2])

            
    def calcular(self):
        nombre=self.entry17.get()
        cant=self.entry25.get() 
        total=int(float(self.entry22.get()))*int(float(self.entry24.get())) #cantidad 
        self.entry25.insert(0, total) 
        
        
    def recargo(self):
        nombre=self.entry17.get()
        recarga=self.entry27.get()
        cursor=cursorObj.execute("select saldo from cuenta where nombre=?", (nombre, ))
        fila1=cursor.fetchone()
        saldoF=int(float(fila1[0]))+int(float(recarga))
        update = """Update cuenta set saldo = ? where nombre = ?"""
        data = (saldoF, nombre)
        cursorObj.execute(update, data) 
        con.commit()
        MessageBox.showwarning("Alerta", "Se ha realizado una recarga")
                
        print("se ha realizado una recarga")

    def comprar(self):
        nombre=self.entry17.get()
        cant=self.entry24.get() 
        total=int(float(self.entry22.get()))*int(float(self.entry24.get())) #cantidad 
        
        cursor=cursorObj.execute("select saldo from cuenta where nombre=?", (nombre, ))
        fila1=cursor.fetchone()
        print(self.entry16.get()+"    "+cant+" "+fila1[0])
        if fila1[0]>str(total):
            if self.entry23.get()>self.entry24.get():
                print(fila1)
                self.entry26.insert(0, int(fila1[0])-total)
                act=self.entry26.get()
                update = """Update cuenta set saldo = ? where nombre = ?"""
                data = (act, nombre)
                cursor.execute(update, data) 
                con.commit() 
            else:
                MessageBox.showwarning("Alerta", "No hay en stock.")
                print("NO hay en stock por cuarentena.")
        else:
            MessageBox.showwarning("Alerta", "No tiene saldo.")
            print("No tiene saldo che pobre.")
        
        
        
        
    def __init__(self):
        self.ventana1=tk.Tk()
        self.ventana1.title("PRUEBA ")
        self.cuaderno1 = ttk.Notebook(self.ventana1)
        
        self.pagina1 = ttk.Frame(self.cuaderno1)
        self.cuaderno1.add(self.pagina1, text="Persona")
        self.label1=ttk.Label(self.pagina1, text="Nombre")
        self.label1.grid(column=0, row=0)
        self.label1=ttk.Label(self.pagina1, text="Direccion")
        self.label1.grid(column=0, row=1)
        self.label1=ttk.Label(self.pagina1, text="Telefono")
        self.label1.grid(column=0, row=2)
        self.label1=ttk.Label(self.pagina1, text="Email")
        self.label1.grid(column=0, row=3)
        self.boton1=ttk.Button(self.pagina1, text="Guardar cliente",command=self.cliente)
        self.boton1.grid(column=2, row=1)
        self.boton2=ttk.Button(self.pagina1, text="Cancelar cliente")
        self.boton2.grid(column=2, row=2) 
        #cajas texto
        self.entry=tk.Entry(self.pagina1, width=30)
        self.entry.grid(column=1, row=0)
        self.entry1=tk.Entry(self.pagina1, width=30)
        self.entry1.grid(column=1, row=1)
        self.entry2=tk.Entry(self.pagina1, width=30)
        self.entry2.grid(column=1, row=2)
        self.entry3=tk.Entry(self.pagina1, width=30)
        self.entry3.grid(column=1, row=3)
        

#ventana 2
        self.pagina2 = ttk.Frame(self.cuaderno1)
        self.cuaderno1.add(self.pagina2, text="Saldo")
        
        self.label1=ttk.Label(self.pagina2, text="Saldo")
        self.label1.grid(column=0, row=0)
        self.label1=ttk.Label(self.pagina2, text="Tarjeta de Credito")
        self.label1.grid(column=0, row=1)
        self.label1=ttk.Label(self.pagina2, text="Nombre CLiente")
        self.label1.grid(column=0, row=2) 
        
        self.boton7=ttk.Button(self.pagina2, text="Guardar Cuenta",command=self.cuenta)
        self.boton7.grid(column=2, row=1)
        self.boton8=ttk.Button(self.pagina2, text="Cancelar Cuenta")
        self.boton8.grid(column=2, row=2) 
        #cajs texto
        self.entry4=tk.Entry(self.pagina2, width=30)
        self.entry4.grid(column=1, row=0)
        self.entry5=tk.Entry(self.pagina2, width=30)
        self.entry5.grid(column=1, row=1)
        self.entry6=tk.Entry(self.pagina2, width=30)
        self.entry6.grid(column=1, row=2)
#ventana 3        
        self.pagina3 = ttk.Frame(self.cuaderno1)
        self.cuaderno1.add(self.pagina3, text="Producto")
 
        self.label1=ttk.Label(self.pagina3, text="Nombre Producto")
        self.label1.grid(column=0, row=0)
        self.label1=ttk.Label(self.pagina3, text="Precio Producto")
        self.label1.grid(column=0, row=1)
        self.label1=ttk.Label(self.pagina3, text="Stock Producto ")
        self.label1.grid(column=0, row=2) 
        
        self.boton12=ttk.Button(self.pagina3, text="Guardar Producto",command=self.producto)
        self.boton12.grid(column=2, row=1)
        self.boton13=ttk.Button(self.pagina3, text="Cancelar Producto")
        self.boton13.grid(column=2, row=2) 
        
        self.entry14=tk.Entry(self.pagina3, width=30)
        self.entry14.grid(column=1, row=0)
        self.entry15=tk.Entry(self.pagina3, width=30)
        self.entry15.grid(column=1, row=1)
        self.entry16=tk.Entry(self.pagina3, width=30)
        self.entry16.grid(column=1, row=2)


#ventana 4       
        self.pagina4 = ttk.Frame(self.cuaderno1)
        self.cuaderno1.add(self.pagina4, text="Pedidos")
        
        self.label1=ttk.Label(self.pagina4, text="Nombre")
        self.label1.grid(column=0, row=0)
        self.label1=ttk.Label(self.pagina4, text="Direccion")
        self.label1.grid(column=0, row=1)
        self.label1=ttk.Label(self.pagina4, text="Telefono")
        self.label1.grid(column=0, row=2)
        self.label1=ttk.Label(self.pagina4, text="Email")
        self.label1.grid(column=0, row=3)
        
        self.entry17=tk.Entry(self.pagina4, width=30)
        self.entry17.grid(column=1, row=0)
        self.entry18=tk.Entry(self.pagina4, width=30)
        self.entry18.grid(column=1, row=1)
        self.entry19=tk.Entry(self.pagina4, width=30)
        self.entry19.grid(column=1, row=2)
        self.entry20=tk.Entry(self.pagina4, width=30)
        self.entry20.grid(column=1, row=3)
 
        self.label1=ttk.Label(self.pagina4, text="Nombre Producto")
        self.label1.grid(column=0, row=4)
        self.label1=ttk.Label(self.pagina4, text="Precio Producto")
        self.label1.grid(column=0, row=5)
        self.label1=ttk.Label(self.pagina4, text="Stock")
        self.label1.grid(column=0, row=6) 
        
        self.entry21=tk.Entry(self.pagina4, width=30)
        self.entry21.grid(column=1, row=4)
        self.entry22=tk.Entry(self.pagina4, width=30)
        self.entry22.grid(column=1, row=5)
        self.entry23=tk.Entry(self.pagina4, width=30)
        self.entry23.grid(column=1, row=6)
        
        self.label1=ttk.Label(self.pagina4, text="Cantidad")
        self.label1.grid(column=0, row=7)
        self.label1=ttk.Label(self.pagina4, text="Total ")
        self.label1.grid(column=0, row=8) 
        self.label1=ttk.Label(self.pagina4, text="Saldo")
        self.label1.grid(column=0, row=9) 
        self.label1=ttk.Label(self.pagina4, text="Recargar Saldo")
        self.label1.grid(column=0, row=10) 
        
        self.entry24=tk.Entry(self.pagina4, width=30)
        self.entry24.grid(column=1, row=7)
        self.entry25=tk.Entry(self.pagina4, width=30)
        self.entry25.grid(column=1, row=8)
        self.entry26=tk.Entry(self.pagina4, width=30)
        self.entry26.grid(column=1, row=9)
        self.entry27=tk.Entry(self.pagina4, width=30)
        self.entry27.grid(column=1, row=10)

        self.boton27=ttk.Button(self.pagina4, text="Buscar Producto",command=self.Bproducto)
        self.boton27.grid(column=2, row=1)
        self.boton28=ttk.Button(self.pagina4, text="Buscar Cliente",command=self.Bcliente)
        self.boton28.grid(column=2, row=2) 
        self.boton29=ttk.Button(self.pagina4, text="Recargar",command=self.recargo)
        self.boton29.grid(column=2, row=3) 
        self.boton30=ttk.Button(self.pagina4, text="Calcular",command=self.calcular)
        self.boton30.grid(column=2, row=4) 
        self.boton31=ttk.Button(self.pagina4, text="Comprar",command=self.comprar)
        self.boton31.grid(column=2, row=5) 


        
        self.cuaderno1.grid(column=0, row=0)        
        self.ventana1.mainloop()
        


aplicacion1=Aplicacion()
```
